```python
N=5
x=[23,43,50,35,68]
y=[25,20,20,48,18]
xe=[2,2,2,2,2]
ye=[4,3,1,2,3]
w=.55                                      #thickness
s=np.arange(N)                             # num of points
p1=plt.bar(s,x,w,yerr=xe)
p2=plt.bar(s,y,w,bottom=x,yerr=ye)         #takes values of x at bottom
plt.xticks(s,('G1','G2','G3','G4','G5'))   # label to x axis
plt.yticks(np.arange(10,100,10)) 
plt.legend((p1,p2),('men','women'))  # describing the graph
```
